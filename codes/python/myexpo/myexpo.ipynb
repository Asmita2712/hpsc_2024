```python
def myexpo(x, n_terms=100):
    result = 0
    term = 1
    prev_term = 0  
    for i in range(n_terms):
        result += term
        prev_term = term  
        term *= x / (i + 1)
      
        if abs(term) < 1e-10 * abs(prev_term): 
            break
    return result
```

```python
myexpo(2)
```

```python
myexpo(0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_exp_comparison():
    x_values = np.linspace(0, 100, 1000)
    exp_values_builtin = np.exp(x_values)
    exp_values_custom = [myexpo(x) for x in x_values]

    plt.plot(x_values, exp_values_builtin, label='numpy.exp()')
    plt.plot(x_values, exp_values_custom, label='my_expo()')
    plt.xlabel('x')
    plt.ylabel('exp(x)')
    plt.title('Comparison of exp(x)')
    plt.legend()
    plt.show()
```
